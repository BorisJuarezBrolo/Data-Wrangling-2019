```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
Data_titanic = pd.read_csv('drive-download-20191016T191440Z-001/titanic_MD.csv')
missing = []
```

## Missing Values Report

### Array creado para evaluar los valores faltantes en las columnas del data set

```python
for i in Data_titanic.columns: 
    if pd.isnull(Data_titanic[i]).any():
        missing.append(i)
print(missing)
```

##### Los valores faltantes de la edad, Parch y Embarked se dejan vacios mientras que los valores faltantes en SibSp toman la forma de Na

## Funciones de limpieza

```python
def pairwise_deletion(dataframe, columns): 
    dataframe = dataframe.dropna(subset=columns)
    return dataframe

def imputation_mode(frame, column_name=None):
    dataframe = frame
    if column_name==None:
        for i in range(183):
            dataframe = dataframe.fillna(dataframe.mode().iloc[i])
    else:
        dataframe[[column_name]] = dataframe[[column_name]].fillna(dataframe.mode().iloc[0])
    return dataframe

def outliers_standard(dataframe, column, zscore):
    dataframe = dataframe[(np.abs(stats.zscore(dataframe[[column]])) < zscore).all(axis=1)]
    return dataframe

def imputation_mean(frame, column_name=None):
    dataframe = frame
    if column_name==None:
        for i in range(7):
            dataframe = dataframe.fillna(dataframe.mean().iloc[i])
    else:
        dataframe[[column_name]] = dataframe[[column_name]].fillna(dataframe[[column_name]].mean())
    return dataframe

def listwise_deletion(dataframe):
    dataframe = dataframe.dropna()
    return dataframe

def imputation_median(frame, column_name=None):
    dataframe = frame
    if column_name==None:
        for i in range(11):
            dataframe = dataframe.fillna(dataframe.median().iloc[i])
    else:
        dataframe[[column_name]] = dataframe[[column_name]].fillna(dataframe[[column_name]].median())
    return dataframe

def listwise_deletion_column(dataframe, column):
    dataframe = dataframe.dropna(subset=column)
    return dataframe

```

## Age

#### Para el caso de esta columna se va a utilizar el metodo de imputación basado en el promedio para tomar el rango promedio de edad del data set y no implicar distinciones con sesgos subjetivos como el valor más repetido

#### Para esta columna se eliminan como primer punto los valores exagerados como primer punto para imputar el promedio en segundo lugar

```python
fare = imputation_mean(Data_titanic, 'Fare')
fare = outliers_standard(Data_titanic, 'Fare', 4)
fare
```

# Aplicación de Metodos individualmente

```python
## Mode Imputation 

mode = Data_titanic
for i in missing:
    mode = imputation_mode(mode, i)
mode.head()
```

```python
## Mean Imputation
mean = Data_titanic
for j in missing:
    mean = imputation_mean(mean, j)
mean.head()
```

```python
## Listwise Deletion 
print("Cantidad de filas previo a limpieza:", len(Data_titanic.index))
print("Cantidad de filas limpias:",len(listwise_deletion(Data_titanic).index))
```

```python
## Pairwise Deletion 
print("Cantidad de filas previo a limpieza:", len(Data_titanic.index))
print("Cantidad de filas limpias:",len(pairwise_deletion(Data_titanic, ['Age']).index))
```

```python
## Median Imputation 

median = Data_titanic
for h in missing:
    median = imputation_median(median, h)
median.head()
```

### La metodología que reduce los sesgos implicitos en la data de mejor manera será la eliminación, sin embargo la moda produce la perdida de la menor cantidad de data sustituyendo con los valores más populares los faltantes

#### Los metodos de imputación resultan ser más comprensibles. 

#### Se introduce sesgo al reemplazar por otros subyacentes pero la reducción de 183 a 158 resulta ser de 13.66% del total de observaciones.
